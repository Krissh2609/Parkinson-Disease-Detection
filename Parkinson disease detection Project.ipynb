data preprocessing

```python
df.drop(['name'],axis=1,inplace=True) # removing name from the dataset
```

```python
from sklearn.model_selection import train_test_split # importing train_test_split  module from sklearn
from sklearn.preprocessing import StandardScaler # importing standarscaler module for standardizing the data
```

```python
X=df.drop(['status'],axis=1) # for the training data seperating target varlue 'status'
y=df['status'] # setting target value as test data
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X) # standardize the data for better classification
```

```python
X_train, X_test, y_train, y_test = train_test_split(
 X_scaled, y, test_size=0.3, random_state=101)  # assign the training and testing data 
```

```python
from sklearn.ensemble import RandomForestClassifier # import the randomforestclassifier from sklearn
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42) # apply the classifier with required parameters. 
model.fit(X_train,y_train) # fit the model with X_train and y_train
```

```python
pred=model.predict(X_test) # make predictions upon the testing data 
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix # import merics to get classification report, accuracy score etc
```

```python
print(classification_report(y_test,pred))
print(accuracy_score(y_test,pred)) # print the classification report and accuracy score to get results
```

According to the classification report, 
the model is able to predict whether person has parkinson disease or is healthy accurately 93% .
It doesn' falsely label people whether health or affected.

81% of people are correctly predicted as health and 98% are correctly predicted for parkinson disease, thus the model is giving correct predictions about the condition most of the time

For class 1 -Parkinson disease, f1-score is 95% thus good balance between precison and recall.

Overall model accuracy is 93% which is pretty good accuracy for predicting whether person is affected or health.

```python
sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
